```python
# 연도, 판매액 lineplot으로 그리고
# 상품군별로 다른 색상으로 표시

sns.lineplot(data=df_total, x='연도', y='백만원', hue='상품군별')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# 위에 그린 그래프를 자세히 보기위해
# 서브플롯으로 표시하기

sns.relplot(data=df_total, x='연도', y='백만원',hue='상품군별'
           , kind='line', col='상품군별', col_wrap=4)
```

```python
# isin을 사용해 
# 화장품 제외
# df_sub라는 변수에 담는다

df_total['상품군별'].isin(['화장품'])
df_sub = df_total[~df_total['상품군별'].isin(['화장품'])].copy()
```

```python
# 연도별 판매액을 상품군별로 
# replot을 활용해 서브플롯으로 그려보기

sns.replot(data=df+sub, x='연도', y='백만원', hue='상품군별'
          , col='상품군별', col_wrap=4, kind='line')

# kind를 붙이지 않으면 점처럼 박힌다
```
